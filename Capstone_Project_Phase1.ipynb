```python
import selenium
import pandas as pd
```

```python
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
#Launching Browser
driver = webdriver.Chrome()
```

```python
driver.get(" https://www.olx.in/")
```

```python
search= driver.find_element(By.XPATH,'/html/body/div/div/div[1]/div/div/div[1]/div[2]/div[4]/a/span')
search.click()
```

```python
#Extracting Urls of different properties at different locations
url0= []
for i in range(5):
    link=driver.find_elements(By.XPATH, '//ul[@class="_266Ly _10aCo"]/li/a')
    for j in link:
        url0.append(j.get_attribute('href'))
    
    
    
url0
```

```python
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException 
```

```python
#Extracting required entries from each property
bhk=[]
p_type=[]
beds=[]
bath=[]
address=[]
landmark=[]

carpetarea=[]
status=[]
floor=[]
transaction_type=[]
yearbuilt=[]
price_sqft=[]

furnished_sts=[]
facing=[]
car=[]
owner=[]
amount_sqft=[]
amount_total=[]


for i in url0:
    driver.get(i)
    time.sleep(5)
    try:
        bh=driver.find_element(By.XPATH, '/html/body/div[1]/div/main/div/div/div/div[3]/section/div/div/div[1]/div/div[2]/div/span[2]')
        bhk.append(bh.text)
    except NoSuchElementException:
        bhk.append('NONE')
    try:
        p_typ=driver.find_element(By.XPATH, '/html/body/div[1]/div/main/div/div/div/div[3]/section/div/div/div[1]/div/div[1]/div/span[2]')
        p_type.append(p_typ.text)
    except NoSuchElementException:
        p_type.append('NONE')
    try:
        bed=driver.find_element(By.XPATH, '/html/body/div[1]/div/main/div/div/div/div[3]/section/div/div/div[1]/div/div[2]/div/span[2]')
        beds.append(bed.text)
    except NoSuchElementException:
        beds.append('NONE') 
    try:
        bat=driver.find_element(By.XPATH, '/html/body/div[1]/div/main/div/div/div/div[3]/section/div/div/div[1]/div/div[3]/div/span[2]')
        bath.append(bat.text)
    except NoSuchElementException:
        bath.append('NONE')
    try:
        add=driver.find_element(By.XPATH, '/html/body/div[1]/div/main/div/div/div/div[4]/div[1]/div/section/div/div[1]/div/span')
        address.append(add.text)
    except NoSuchElementException:
        address.append('NONE')
    try:
        land=driver.find_element(By.XPATH, '/html/body/div[1]/div/main/div/div/div/div[3]/section/div/div/div[1]/div/div[14]/div/span[2]')
        landmark.append(land.text)
    except NoSuchElementException:
        landmark.append('NONE')
    try:
        carpet=driver.find_element(By.XPATH, '/html/body/div[1]/div/main/div/div/div/div[3]/section/div/div/div[1]/div/div[8]/div/span[2]')
        carpetarea.append(carpet.text)
    except NoSuchElementException:
        carpetarea.append('NONE')
    try:
        stat=driver.find_element(By.XPATH, '/html/body/div[1]/div/main/div/div/div/div[3]/section/div/div/div[1]/div/div[5]/div/span[2]')
        status.append(stat.text)
    except NoSuchElementException:
        status.append('NONE')
    try:
        floo=driver.find_element(By.XPATH, '/html/body/div[1]/div/main/div/div/div/div[3]/section/div/div/div[1]/div/div[11]/div/span[2]')
        floor.append(floo.text)
    except NoSuchElementException:
        floor.append('NONE')
    try:
        pri=driver.find_element(By.XPATH, '/html/body/div[1]/div/main/div/div/div/div[3]/section/div/div/div[1]/div/div[7]/div/span[2]')
        price_sqft.append(pri.text)
    except NoSuchElementException:
        price_sqft.append('NONE')
    try:
        furnished_st=driver.find_element(By.XPATH, '/html/body/div[1]/div/main/div/div/div/div[3]/section/div/div/div[1]/div/div[4]/div/span[2]')
        furnished_sts.append(furnished_st.text)
    except NoSuchElementException:
        furnished_sts.append('NONE')
    try:
        facin=driver.find_element(By.XPATH, '/html/body/div[1]/div/main/div/div/div/div[3]/section/div/div/div[1]/div/div[13]/div/span[2]')
        facing.append(facin.text)
    except NoSuchElementException:
        facing.append('NONE')
        
    try:
        ca=driver.find_element(By.XPATH, '/html/body/div[1]/div/main/div/div/div/div[3]/section/div/div/div[1]/div/div[12]/div/span[2]')
        car.append(ca.text)
    except NoSuchElementException:
        car.append('NONE')
    try:
        own=driver.find_element(By.XPATH, '/html/body/div[1]/div/main/div/div/div/div[3]/section/div/div/div[1]/div/div[6]/div/span[2]')
        owner.append(own.text)
    except NoSuchElementException:
        owner.append('NONE')
    try:
        amt=driver.find_element(By.XPATH, '/html/body/div[1]/div/main/div/div/div/div[3]/section/div/div/div[1]/div/div[10]/div/span[2]')
        amount_sqft.append(amt.text)
    except NoSuchElementException:
        amount_sqft.append('NONE')       
    try:
        am=driver.find_element(By.XPATH, '/html/body/div/div/main/div/div/div/div[4]/div[1]/div/section/span[1]')
        amount_total.append(am.text)
    except NoSuchElementException:
        amount_total.append('NONE') 
```

```python
import pandas as pd
df1= pd.DataFrame({'Bhk_Type':bhk ,'Property_Type':p_type,'Beds':beds,'Bathroom':bath,'Address':address,'Landmark':landmark})
df2=pd.DataFrame({'Carpet_Area':carpetarea,'Status':status,'Floors':floor,'Price_sqft':price_sqft})
df3=pd.DataFrame({'Furnishing':furnished_sts,'Facing':facing,'Car_Parking':car,'Ownership':owner,'Book_amt_Lakhs':amount_sqft,'Total_Amt_in_rupees':amount_total})
df1
```

```python
df2['Floors']=df2['Floors'].apply(pd.to_numeric, errors='coerce')
df2
```

```python
df3['Car_Parking']=df3['Car_Parking'].apply(pd.to_numeric, errors='coerce')

#Removing 'Rupee sign ' from amount
df3['Total_Amt_in_rupees'] = df3['Total_Amt_in_rupees'].str.replace('₹', '')
df3
```
